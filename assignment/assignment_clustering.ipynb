````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from keras.datasets import mnist
````

1. Run the code below, which creates four datasets: `df0_125`, `df0_25`, `df0_5`, `df1_0`, and `df2_0`. Each data set is created by increasing the amount of `noise` (standard deviation) around the cluster centers, from `0.125` to `0.25` to `0.5` to `1.0` to `2.0`.

````python
def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?

````python
datasets = [
    (df0_125, "noise=0.125"),
    (df0_25,  "noise=0.25"),
    (df0_5,   "noise=0.5"),
    (df1_0,   "noise=1.0"),
    (df2_0,   "noise=2.0")
]

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(30, 6), sharex=True, sharey=True)

for ax, (df, title) in zip(axes, datasets):
    for grp in df['group'].unique():
        subset = df[df['group'] == grp]
        ax.scatter(subset['x1'], subset['x2'], label=grp)
    ax.set_title(title)
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")

plt.tight_layout()
plt.show()
````

As the noise increases, the visual distincitveness of the clusters decreases greatly as the points are more spread out and less concentrated around the cluster centers.

3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$. (Pay attention to the vertical axis across plots, or put all the scree curves on a single canvas.)

````python
k_values = range(1, 8)

plt.figure(figsize=(8, 6))

for df, label in datasets:
    sse = []
    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=100)
        kmeans.fit(df[["x1", "x2"]])
        sse.append(kmeans.inertia_)
    plt.plot(k_values, sse, marker='o', label=label)

plt.xlabel("Number of clusters (k)")
plt.ylabel("SSE")
plt.title("Scree Plot Comparison for Different Noise Levels")
plt.legend()
plt.grid(True)
plt.show()

````

As the level of k increases, the scree plot becomes less apparent because the elbow becomes far less defined. This makes it harder to select a k value as the noise increases.

4. Explain the intuition of the elbow, using this numerical simulation as an example.

The elbow in a scree plot represents the point in which increasing the number of clusters does not decrease the sse (sum of squared errors) by a significant amount. The elbow is important when selecting the number of clusters because it shows when the clusters are no longer distinct. The elbow is easier to see when the noise is lower because the clusters are more distinct, meaning the sse is lower. As the noise increases, the clusters become less distinct, making it harder to see the elbow in the scree plot.

5. Cross tabulate the cluster assignments with the true labels for the test set values. How good is the correspondence? What proportion of digits are clustered correctly? Which digits are the hardest to distinguish from one another? Can $k$MC recover the latent digits 0 to 9, without even knowing what those digits were?

````python
cross_tab = pd.crosstab(y_test, cluster_labels, rownames=['True Label'], colnames=['Cluster'])

plt.figure(figsize=(8, 6))
sns.heatmap(cross_tab, annot=True, fmt="d")
plt.title("Confusion Matrix: True Label vs. Cluster")
plt.show()

cluster_mapping = cross_tab.idxmax(axis=0)
mapped_predictions = np.array([cluster_mapping[cl] for cl in cluster_labels])

accuracy = accuracy_score(y_test, mapped_predictions)
print(f"Overall clustering accuracy: {accuracy:.2%}")

misclassified = (y_test != mapped_predictions)
print("Most misclassified digits:")
print(pd.Series(y_test[misclassified]).value_counts().head())
````

Based on the scree plot, the elbow appears around k = 5 or 6 which means that the algorithm would not pick 10 clusters. The algorithm tends to combine digits that are similar in shape into the same classification. For example, 4 and 9 are often combined into the same classification because they have similar shapes.
